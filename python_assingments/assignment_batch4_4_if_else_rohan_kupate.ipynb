### 36. Determine the type of triangle using sides: acute, right, obtuse.

```python
a1 = 90
a2 = 60
a3 = 30

if a1 + a2 + a3 == 180 and a1>0 and a2>0 and a3>0:
    if a1 == a2 == a3:
        print("Equilateral triangle")
    elif a1 == 90 or a2 == 90 or a3 == 90:
        print("Right angled triangle")
    elif a1 < 90 and a2 < 90 and a3 < 90:
        print("Acute angled triangle")
    elif a1 > 90 or a2 > 90 or a3 > 90:
        print("Obtuse angled triangle")
    else:
        print("Scalene triangle")
else:
    print("Not a valid triangle")
```

### 42. Check if a number is perfect (sum of factors equals number).

```python
n42 = 6
factors_of_n42 = []

for i in range(1, n42):
    if n42 % i == 0:
        factors_of_n42.append(i)

# print(n42, "is a perfect number") if sum(factors_of_n42) == n42 else print(n42, "is not a perfect number")

s42 = sum(factors_of_n42)
if s42==n42:
    print(n42,"is a perfect number")
else:
    print(n42,"is not a perfect number")
   
```
