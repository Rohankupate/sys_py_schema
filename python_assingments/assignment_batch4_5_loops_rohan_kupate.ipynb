### 25. Given a dictionary, find the key with the highest value using a for loop

```python
student_dict = {'Alice': 85, 'Bob': 92, 'Charlie': 78, 'Diana': 96}
topper = {}
topper_m= 0

for i in student_dict:
    if student_dict[i] > topper_m:
        if not topper.keys():
            topper[i] = student_dict[i]

        else:
            topper.clear()
            topper[i] = student_dict[i]

        topper_m = student_dict[i]
print("Topper:", topper)
```
